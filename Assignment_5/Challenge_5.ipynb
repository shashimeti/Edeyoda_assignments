```python
class Account:
    
    def __init__(self, title=None, balance=0):
        self.title = title
        self.balance = balance
    
    def withdrawal(self, amount):
        # write code here
        if amount > 0 and amount <= self.balance:
            self.balance -=amount

    def deposit(self, amount):
        # write code here
        if amount > 0:
            self.balance += amount
    def getBalance(self):
        # write code here
        return self.balance

class SavingsAccount(Account):
    def __init__(self, title=None, balance=0, interestRate=0):
            super().__init__(title, balance)
            self.interestRate = interestRate
    
    def interestAmount(self):
        # write code here
        InterstAmount = (self.interestRate * self.balance)/100
        return InterstAmount

#code to test - do not edit this

demo1 = SavingsAccount("Ashish", 2000, 5)   # initializing a SavingsAccount object

print("Account Name: ",demo1.title)
print("Account Balance: ",demo1.balance)

demo1.deposit(500)
print("-----------Deposit----------")
print("Balance after Deposite: ",demo1.getBalance())

demo1.withdrawal(500)
print("-----------Withdrawal-----------")
print("After withdrawal balance: ",demo1.getBalance())

SA = SavingsAccount("Ashish",2000,5)
print("-----------Savings Account-------------")
print("Initial Savings Balance Amount: ",SA.getBalance())

Interest = SA.interestAmount()
print("------------Interest Amount------------")
print("Interest Amount: ",Interest)


```
