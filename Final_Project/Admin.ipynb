ðŸ‘‰ 1. Add new food items. Food Item will have the following details:
        ðŸ”´ FoodID //It should be generated automatically by the application.
        ðŸ”´ Name
        ðŸ”´ Quantity. For eg, 100ml, 250gm, 4pieces etc
        ðŸ”´ Price
        ðŸ”´ Discount
        ðŸ”´ Stock. Amount left in stock in the restaurant.

ðŸ‘‰ 2. Edit food items using FoodID.

ðŸ‘‰ 3. View the list of all food items.

ðŸ‘‰ 4. Remove a food item from the menu using FoodID.

```python
class FoodItem:
    def __init__(self,Name,Quantity,Price,Discount,Stock) -> None:
        
        self.Name = Name
        self.Quantity = Quantity
        self.Price = Price
        self.Discount = Discount
        self.Stock = Stock

class Admin:
    def __init__(self) -> None:
        self.food_items = {}
        self.food_id_counter = 1

    def add_food_item(self,Name,Quantity,Price,Discount,Stock):
        FoodID = self.food_id_counter
        self.food_id_counter+=1
        self.food_items[FoodID] = FoodItem(Name,Quantity,Price,Discount,Stock)
        print("Food items added succesfull")

    def edit_food_item(self,FoodID,Name=None,Quantity=None,Price=None,Discount=None,Stock=None):
        if FoodID in self.food_items:
            Food_item = self.food_items[FoodID]

            if Name:
                Food_item.Name = Name
            if Quantity:
                Food_item.Quantity = Quantity
            if Price:
                Food_item.Price = Price
            if Discount:
                Food_item.Discount = Discount
            if Stock:
                Food_item.Stock = Stock
            print("Items Edited succesfully.")
        else:
            print("Food Item with FoodID", FoodID, "not found")

    def view_food_items(self):
        print("List of all food items:")
        for FoodID, Food_item in self.food_items.items():
            print("FoodID:", FoodID)
            print("Name:", Food_item.Name)
            print("Quantity:", Food_item.Quantity)
            print("Price:", Food_item.Price)
            print("Discount:", Food_item.Discount)
            print("Stock:", Food_item.Stock)
            print()

    def remove_food_item(self, FoodID):
        if FoodID in self.food_items:
            del self.food_items[FoodID]
            print("Food item with FoodID", FoodID, "removed successfully.")
        else:
            print("Food item with FoodID", FoodID, "not found.")

# Sample usage
admin = Admin()

# Adding food items
admin.add_food_item("Burger", "1 piece", 150, 10, 50)
admin.add_food_item("Pizza", "Large", 300, 15, 30)

# Viewing food items
admin.view_food_items()

# Editing a food item
admin.edit_food_item(1, Price=160, Stock=45)

# Viewing updated food items
admin.view_food_items()

# Removing a food item
admin.remove_food_item(2)

# Viewing updated food items
admin.view_food_items()

```
